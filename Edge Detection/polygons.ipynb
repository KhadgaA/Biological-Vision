```python
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon, Circle
```

```python



def Create_polygon(points:list[tuple[float,float]], name:str ="polygon", saveimg:bool = False)-> None:
    fig,ax = plt.subplots(figsize = (5,5))
    polygon = Polygon(points,fill = False)
    ax.add_patch(p=polygon)
    plt.xlim(0,5)
    plt.ylim(0,5)
    plt.axis(False)
    plt.plot()
    if saveimg:
        fig.savefig(name)

def Create_circle(center:tuple[float,float], radius:float, name:str ="Circle", saveimg:bool = False)-> None:
    fig,ax = plt.subplots(figsize = (5,5))
    polygon = Circle(xy = center, radius=radius,fill = False)
    ax.add_patch(p=polygon)
    plt.xlim(0,5)
    plt.ylim(0,5)
    plt.axis(False)
    plt.plot()
    if saveimg:
        fig.savefig(name)
```

```python
Create_polygon([(1,1),(3,1),(2,3)],saveimg=True, name = 'Triangle')
```

```python
Create_polygon([(1,1),(3,1),(3,4),(1,4)],saveimg=True, name = 'Rectangle')
```

```python
Create_polygon([(1,1),(1,2),(2,3),(3,2),(3,1)],saveimg=True, name = 'Pentagon')
```

```python
Create_circle((2.5,2.5),2,saveimg=True, name = 'Circle')
```
